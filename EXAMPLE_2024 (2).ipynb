```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sci
```

```python
#исходные данные задачи 
##### PART 1
n=120
mu, sigma = 0,16 
# theoretical mean and standard deviation
```

```python
s = np.random.normal(mu, sigma, n)
a_estimate=np.mean(s) #среднее
sigma_estimate=np.std(s,ddof=1)#ddof - дельта спепеней свободы
(a_estimate,sigma_estimate)
print(s)
sum=0
for i in s:
    sum+=(i-a_estimate)**2
sum/= n-1
sum=math.sqrt(sum)
print(sum)
print(sigma_estimate)
```

```python
count, bin_edges = np.histogram(s,4)
#count, bin_edges = np.histogram(s,k_bins)#k_bins - 
### count содержит абсолютные частоты для каждого интервала
print(count)
print(bin_edges)
count1=np.sum(count)
count1
```

```python
count, bin_edges = np.histogram(s,4,density=True)
### count содержит абсолютные частоты для каждого интервала
print(count)
print(bin_edges)

count, bins, ignored  = plt.hist(s, 4, density=True)
### гистограмма частот: сумма высот прямоугольников равна объему выборки
### гистограмма относительных частот: сумма площадей  прямоугольников равна 1

shag=bins[1]-bins[0]
count2=np.sum(count*shag)
count2
```

```python
count, bin_edges = np.histogram(s,250)
### bin_edges - абсциссы точек, для которых находим значения функции распределения и затем плотности
from scipy.stats import norm
plt.plot(bin_edges, norm.cdf(bin_edges,33,1),linewidth=2, color='g')
```

```python
plt.plot(bin_edges, norm.pdf(bin_edges,33,1),linewidth=2, color='r')
count, bins, ignored  = plt.hist(s, 4, density=True)
```

```python
k_bins=4
count, bins, ignored  = plt.hist(s, k_bins)
### гистограмма частот: сумма высот прямоугольников равна объему выборки
np.sum(count)
```

```python
k_bins=4
count, bins, ignored  = plt.hist(s, k_bins,density="True")
### гистограмма относительных частот: сумма площадей  прямоугольников равна 1
print(count)
print(bin_edges)
np.sum(count)
shag=bins[1]-bins[0]
np.sum(count*shag)
```

```python
count, bins, ignored  = plt.hist(s, bins='fd',density="True")
### гистограмма относительных частот: сумма площадей  прямоугольников равна 1
## число интервалов группировки найдено методом Фридмана-Диакониса
print(count)
print(bins)
np.sum(count)
shag=bins[1]-bins[0]
np.sum(count*shag)
```
