```python
def householder_reflection(v):
    v_ = v[3:]  # часть вектора с 4 по 10 элемент
    e = np.zeros_like(v_)
    e[0] = np.linalg.norm(v_)
    u = v_ - e
    u = u / np.linalg.norm(u)
    H = np.eye(len(v_)) - 2 * np.outer(u, u)
    v[3:] = H @ v_
    return v

vector_householder = householder_reflection(vector.copy())
print(f"Отражение Хаусхолдера: {vector_householder}\n")
```

# 6. Для матрицы, созданной в 4 пункте, найти LU разложение самостоятельно написанной функцией и проверить результат с помощью перемножения матриц.

```python
def lu_decomposition(matrix):
    n = len(matrix)
    L = np.eye(n)
    U = matrix.copy()
    
    for i in range(n):
        for j in range(i+1, n):
            factor = U[j, i] / U[i, i]
            L[j, i] = factor
            U[j, i:] = U[j, i:] - factor * U[i, i:]
    
    return L, U

L_manual, U_manual = lu_decomposition(matrix_random)

P, L_check, U_check = lu(matrix_random)

print(f"L (вручную):\n{L_manual}\n")
print(f"U (вручную):\n{U_manual}\n")
print(f"L (SciPy):\n{L_check}\n")
print(f"U (SciPy):\n{U_check}\n")
```

# 7. Для матрицы, созданной в 4 пункте, найти QR разложение всеми рассмотренными методами, проверить результат с помощью перемножения матриц и с помощью функции np.linalg.qr

```python
def qr_gram_schmidt(A):
    n, m = A.shape
    Q = np.zeros((n, m))
    R = np.zeros((m, m))
    
    for j in range(m):
        v = A[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], A[:, j])
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]
    
    return Q, R

Q_gs, R_gs = qr_gram_schmidt(matrix_random)

Q_check, R_check = qr(matrix_random)

print(f"Q (Грам-Шмидт):\n{Q_gs}\n")
print(f"R (Грам-Шмидт):\n{R_gs}\n")
print(f"Q (SciPy):\n{Q_check}\n")
print(f"R (SciPy):\n{R_check}\n")
```

# 1. Создать произвольную верхнетреугольную матрицу А 5 порядка (не унитреугольную), вектор B произвольный. Решить систему AX = B методом обратной подстановки. Проверить решение методом np.solve

```python
# Создаем верхнетреугольную матрицу A размером 5x5
A = np.array([
    [2, -1, 0, 0, 0],
    [0, 3, 1, 0, 0],
    [0, 0, 4, 2, 0],
    [0, 0, 0, 5, 1],
    [0, 0, 0, 0, 6]
])

# Создаем произвольный вектор B
B = np.array([1, 2, 3, 4, 5])

# Метод обратной подстановки
def back_substitution(A, B):
    n = len(B)
    X = np.zeros(n)
    for i in range(n-1, -1, -1):
        X[i] = (B[i] - np.dot(A[i, i+1:], X[i+1:])) / A[i, i]
    return X

# Решаем систему AX = B
X_custom = back_substitution(A, B)

# Проверяем решение с помощью np.solve
X_np = np.linalg.solve(A, B)

# Выводим результаты
print("Решение методом обратной подстановки:", X_custom)
print("Решение с помощью np.solve:", X_np)
print("Совпадают ли решения?:", np.allclose(X_custom, X_np))
```

# 3. Решить систему из пункта 2 с помощью QR разложения матрицы А. QR разложение найти методом Хаусхолдера. Проверить полученное решение непосредственной подстановкой в исходную систему, а также методом np.solve

```python
# Определяем матрицу A и вектор B
A = np.array([
    [4.4, -2.5, 19.2, -10.8],
    [5.5, -9.3, -14.2, 13.2],
    [7.1, -11.5, 5.3, -6.7],
    [14.2, 23.4, -8.8, 5.3]
])

B = np.array([4.3, 6.8, -1.8, 7.2])

# Выполняем QR-разложение
Q, R = qr(A)

# Решаем систему Rx = Q^T * B
QT_B = np.dot(Q.T, B)
X = np.linalg.solve(R, QT_B)

# Проверяем решение, подставляя его в исходную систему
check = np.allclose(np.dot(A, X), B)

# Также решаем систему с помощью np.solve
X_np = np.linalg.solve(A, B)

# Выводим результаты
print("Решение системы уравнений (QR):", X)
print("Проверка (подстановка):", check)
print("Решение с помощью np.solve:", X_np)
print("Совпадают ли решения?:", np.allclose(X, X_np))
```

# 4. Решить систему методом простых итераций с точностью до 10^−3 . 

```python
from prettytable import PrettyTable
```

```python
x1, x2, x3 = 0, 0, 0
tolerance = 1e-3
max_iterations = 100
iterations = []

# Проверяем достаточное условие сходимости (норма по строкам)
matrix_norm = max([abs(3.3 / 2.7) + abs(1.3 / 2.7), abs(3.5 / 1.7) + abs(2.8 / 1.7), abs(4.1 / 1.7) + abs(5.8 / 1.7)])
if matrix_norm >= 1:
    print("Матрица не удовлетворяет условию сходимости. Требуется преобразование системы.")
else:
    print(f"Условие сходимости выполнено, норма матрицы: {matrix_norm:.4f}")

# Метод простых итераций
for i in range(max_iterations):
    x1_new = (2.1 - 3.3 * x2 - 1.3 * x3) / 2.7
    x2_new = (3.5 * x1 + 2.8 * x3 - 1.7) / 1.7
    x3_new = (4.1 * x1 + 5.8 * x2) / 1.7

    # Записываем итерацию
    iterations.append((i + 1, x1_new, x2_new, x3_new))

    # Проверяем сходимость
    if np.allclose([x1_new, x2_new, x3_new], [x1, x2, x3], atol=tolerance):
        break
    x1, x2, x3 = x1_new, x2_new, x3_new

table = PrettyTable()
table.field_names = ["Итерация", "x1", "x2", "x3"]
for iteration in iterations:
    table.add_row(iteration)

print(table)
```
